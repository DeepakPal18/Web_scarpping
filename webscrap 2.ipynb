```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.162 Safari/537.36'
}

webpage = requests.get('https://tax2win.in/guide/gst-hsn-code', headers=headers).text
soup = BeautifulSoup(webpage, "html.parser")

# Fetch links as List of Tag Objects
links=[]
for i in soup.find_all('a')[85:185]:
  links.append(i)

data_list = []
chapter_number = 1
for link in links:
    link_url = link.get('href')
    new_webpage = requests.get(link_url, headers=headers)
    soup_link = BeautifulSoup(new_webpage.content, "html.parser")
    for i in new_soup.find_all('td'):
        print(i.text.strip())
        data_list.append(['Chapter ' + str(chapter_number)])
        data_list.append([i.text.strip()])
        
    chapter_number += 1
```
