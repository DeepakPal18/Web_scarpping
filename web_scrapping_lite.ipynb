```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.162 Safari/537.36'
}

webpage = requests.get('http://rcfcsouthern.org/search.php?id=2', headers=headers).text
soup = BeautifulSoup(webpage, "html.parser")

# Fetch links as List of Tag Objects
links = soup.find_all('a', attrs={'class': 'label label-success'})

data_list = []

for link in links:
    link_url = 'http://rcfcsouthern.org/' + link.get('href')
    new_webpage = requests.get(link_url, headers=headers)
    soup_link = BeautifulSoup(new_webpage.content, "html.parser")

    # Extracting data for each link
    category = soup_link.find_all('td')[1].text.strip()
    name = soup_link.find_all('td')[3].text.strip()
    address = soup_link.find_all('td')[5].text.strip()
    pincode = soup_link.find_all('td')[9].text.strip()
    district = soup_link.find_all('td')[11].text.strip()
    state = soup_link.find_all('td')[13].text.strip()
    mobile = soup_link.find_all('td')[15].text.strip()

    data_list.append({
        'category': category,
        'name': name,
        'address': address,
        'pincode': pincode,
        'district': district,
        'state': state,
        'mobile': mobile,
    })

# Convert list of dictionaries to DataFrame
final_df = pd.DataFrame(data_list)
final_df
```
